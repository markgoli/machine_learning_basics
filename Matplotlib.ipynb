```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 3
```

```python
#DPI - DOTS PER INCH

plt.plot(x,y,'r-')
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('Values')
plt.show()
```

```python
# syntax subplot(rows, columns, plot_number )
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## OBJECT ORIENTED METHOD.

fig = plt.figure() #creating a figure canvas or object 
axes = fig.add_axes([0.1,0.1,0.8,0.8])  # left of the axis, bottom, width , height of the axis
axes.plot(x,y) # plotting the figure
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
# method 2 
fig = plt.figure() #creating a figure canvas or object 
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('Outer plot')

axes2.plot(y,x)
axes2.set_xlabel('x label')
axes2.set_ylabel('y label')
axes2.set_title('Inner Plot')
```

```python
# creating subplots using OO

fig, axes = plt.subplots(nrows=1, ncols=2)
# axes.plot(x,y)
# plt.tight_layout()
# iterating through the axex
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
 # Option 2
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].plot(y,x)
axes[1].set_title('Plot 2')
```

# FIGURE SIZE, ASPECT RATIO AND DOTS PER INCH(DPI)

```python
fig = plt.figure(figsize=(8,2), dpi=1000) # 8 * 2 inches in figure_size
ax = fig.add_axes([0.1,0.2,0.8,0.8])
```

# Legends 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='x squared')
ax.plot(x, x **3, label='x cubed')
ax.legend()
```
